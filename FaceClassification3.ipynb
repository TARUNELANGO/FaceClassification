```python
image=cv2.imread("C:\\Users\\Tarun\\UF3\\D\\test\\Nihaal\\e069ae01-fde7-11ed-aac7-843451f97ba8.jpg")
image = cv2.resize(image, (224, 224))
image = np.expand_dims(image, axis=0)
results=faceClassifier.predict(image)
results=results.reshape(2)
results
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    _ , frame = cap.read()
    frame = frame[50:500, 50:500,:]
    
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    resized = tf.image.resize(rgb, (224,224))
    results=faceClassifier.predict(np.expand_dims(resized,axis=0))
    results=results.reshape(2)
    max=results[0]
    temp=0
    c=0
    while True:
        if results[temp] > max:
                max=results[temp]
                c=temp
        temp=temp+1
        if temp==2:
            break
    if c==0:
        confidence = int( 100 * (1 - (max)/400) )
        display_string = str(confidence) + '% confident it is Nihaal'
        cv2.putText(frame, display_string, (0, 120), cv2.FONT_HERSHEY_COMPLEX, 1, (255,120,150), 1)
        cv2.imshow('EyeTrack',frame)
    elif c==1:
        confidence = int( 100 * (1 - (max)/400) )
        display_string = str(confidence) + '% confident it is Tarun'
        cv2.putText(frame, display_string, (0, 120), cv2.FONT_HERSHEY_COMPLEX, 1, (255,120,150), 1)
        cv2.imshow('EyeTrack',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
